```python
## Predicting attributes (features)
X = records.drop('book_table', axis=1)

# Defining the label
y = records['book_table']

# Displaying the first few rows of the label 'y'
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
## splitting the dataset into training and testing (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state=2023 )
```

```python
## Normalizing all different features
scaler = StandardScaler()

# on the training dataset fitting the scaled...
X_train_scaled = scaler.fit_transform(X_train)

# using the same scaler we are here transforming the data
X_test_scaled = scaler.transform(X_test)
```

```python
## Importing the KNN classifier's classs
from sklearn.neighbors import KNeighborsClassifier

## test train splitting fuction importing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

##for calculating the accuracy
from sklearn import metrics
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
## nearest neighbours??
k = 15

## Euclidean distance (p=2)
knn = KNeighborsClassifier(n_neighbors=k, p=2)

# Fitting the model with the scaled training data
knn.fit(X_train_scaled, y_train)

# Making predictions using the trained KNN classifier on the scaled test data
y_pred = knn.predict(X_test_scaled)
```

```python
## FOr classification : Printing report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
## for the main class printing the forcasted probabilities
y_pred_probs_norm = knn.predict_proba(X_test_scaled)
y_pred_probs_norm = y_pred_probs_norm[:, 1]
print(y_pred_probs_norm)
```

```python
### printing the ROC Curve

from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test, y_pred_probs_norm)
print('AUC:', '%.3f' % metrics.auc(fpr, tpr))
sns.scatterplot(x=fpr, y=tpr, label='Roc Curve')
```

```python
## Evaluation: Model
print("Accuracy: ", '%.3f' % metrics.accuracy_score(y_test,y_pred))
print("Precision: ", '%.3f' % metrics.precision_score(y_test,y_pred))
print("Recall: ", '%.3f' % metrics.recall_score(y_test,y_pred))
print("F1: ", '%.3f' % metrics.f1_score(y_test,y_pred))
```
