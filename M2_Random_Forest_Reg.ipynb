```python
## to splitting the dataset given to us in two, training and testing with the ration of 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
## importing the random forest regressor
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)

## training the predictive model
random_forest_model.fit(X_train, y_train)

##fitting the model
y_pred_rf = random_forest_model.predict(X_test)

## calculating the mean squared value
mse_rf = mean_squared_error(y_test, y_pred_rf)

## printingthe value of mean squared
print("RFR: Mean Squared Error:", mse_rf)

```

```python
from sklearn.metrics import r2_score

##for the calculation of the r-sqrd value of the random forest regressor.
r2_rf = r2_score(y_test, y_pred_rf)

## Printing the R-squared value
print("Random Forest Regressor R-squared:", r2_rf)
```
