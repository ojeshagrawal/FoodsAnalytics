Question 1

```python
#### Que 1.  How many restaurants offer table booking, and what impact does it have on ratings (rate)? using rate and table booking column
```

```python
##counting all the rest. that are having table booking and that no having.
table_booking_counts = records['book_table'].value_counts()
```

```python
## calculating the average "rate" column and classfing it into those that provide table booking and those rest. which dont.
table_booking_ratings = records.groupby('book_table')['rate'].mean()
```

```python
## plotting a graph depicting all those rest. which are offering table bookingss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=table_booking_counts.index, y=table_booking_counts.values)
plt.title('Restaurants (no.) which offer Table Booking')

## finally to visualise how impactful is table booking on the "rate" (ratings) of the rest.
plt.subplot(1, 2, 2)
sns.barplot(x=table_booking_ratings.index, y=table_booking_ratings.values)
plt.title('Table Booking~Ratings impact')
plt.show()
```

```python
## Que 2. How many restaurants offer online ordering, and what impact does it have on ratings (rate)? here we are using
## two columns, "rate" and "online_order"
```

```python
## Number of restr. which offer online ordering (count)
online_count = records['online_order'].value_counts()
```

```python
## average of rate column, when and if restr. offering onling ordering or not
online_ratings = records.groupby('online_order')['rate'].mean()
```

```python
## graph of count (all the restr. having online ordering system)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=online_count.index, y=online_count.values)
plt.title('No. (Restaurants) whichoffer Online Ordering')

## how has online ordering system impacted the ratings of a restr. (rate)
plt.subplot(1, 2, 2)
sns.barplot(x=online_ratings.index, y=online_ratings.values)
plt.title('Impact on Ratings by Online Ordering')
plt.show()
```

```python
## Finding the best rated restraunts (based on the rate column)
# we will sort all the rest. which have highest rating using the columns: "rate" and "name"
best_top_rated_restr = records[['name', 'rate']].sort_values(by='rate', ascending=False).head(10)
## printing the restr. which are best rated
best_top_rated_restr
```

```python
## Finding the least rated restraunts (based on the rate column)
# we will sort all the rest. which have least rating using the columns: "rate" and "name"
worst_rated_restr = records[['name', 'rate']].sort_values(by='rate').head(10)
## printing the restr. which are least rated
worst_rated_restr
```

```python
## city with most number of restr.
## counting the restr.
counts_cty = records['listed_in_city'].value_counts()

## identifying the city with max no. of rest (MAX***)
city_with_most_restaurants = counts_cty.idxmax()

##printing city name
print("City with the most restaurants:", city_with_most_restaurants)
city_with_most_restaurants
```

```python
## city with least number of restr.
## counting the restr.
counts_cty = records['listed_in_city'].value_counts()

## identifying the city with min no. of rest (MIN***)
city_with_least_restaurants = counts_cty.idxmin()

##printing city name
print("City with the least restaurants:", city_with_least_restaurants)
city_with_least_restaurants
```
